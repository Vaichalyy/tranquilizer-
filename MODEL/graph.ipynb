```python
import matplotlib.pyplot as plt
import numpy as np
import pytest

series = []
#read the previous posture status to display the graph
file = open('state_values.txt', 'r') 
while file.readline() != "":
    try:
        series.append(int(file.readline()))
    except:
        pass
file.close()
# calculate read status array list length for adjusting x axis
length = len(series) 
def graphVisualization(x, y):
    x_axis = np.array(range(x))
    y_axis = np.array(y)

    plt.plot(x_axis, y_axis)
    plt.title('Anxiety detection')
    plt.xlabel('seconds')
    plt.ylabel('state\n0 - Anxious, 1 - Neutral')
    return plt.show()
graphVisualization(length, series)
    

def test_graphVisualization():
    x = np.array(range(2))
    y = np.array([0,1])
    plt.plot(x, y)
    assert graphVisualization(x, y) == plt.show()
test_graphVisualization()
```
